```python
# Question 5

marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}

import matplotlib.pyplot as plt
import numpy as np

i = marks.keys()
list1 = []

for k in i:
    list1.append(k)
    
x = np.array(list1)

j = marks.values()
list2 = []

for v in j:
    print(v)
    list2.append(v)

y = np.array(list2)

# Question 5, Bar chart   

plt.title('Student Marks')
plt.bar(x,y)
plt.show()

# Question 5, Pie chart

plt.title('Student Marks')
plt.pie(y, labels = x)
plt.legend(bbox_to_anchor = (1.05, 1.0), loc = 'upper left')
plt.show()
```

```python
# Question 6

import matplotlib.pyplot as plt
import numpy as np

# Subplot with multiple lines

plt.subplot(2,3,1)
plt.title('Subplot with Lines')
x1 = np.array([1,2,3,4])
y1 = np.array([1,2,3,4])
y2 = np.array([5,6,7,8])
y3 = np.array([9,10,11,12])

plt.plot(x1, y1, x1, y2, x1, y3)
plt.show()

# Subplot Bar chart

plt.subplot(2,3,1)
plt.title('Bar Chart')

x = np.array([0.25, 0.5, 0.75])
y = np.array([0.5, 1, 0.3])

plt.bar(x, y, width = 0.1)
plt.show()

# Subplot Pie chart

plt.subplot(2,3,1)
plt.title('Pie Chart')

x = np.array([0.25, 0.5, 0.75])

plt.pie(x)
plt.show()

# Subplot Grid

plt.subplot(2,3,1)
plt.title('Grid')

plt.grid()
plt.show()

# Subplot Scatterplot

plt.subplot(2,3,1)
plt.title('Scatterplot')

x = np.array([0.25, 0.3, 0.5, 0.75])
y = np.array([0.5, 1, 1.7, 2])

plt.scatter(x, y)
plt.show()

# Subplot Histogram

plt.subplot(2,3,1)
plt.title('Histogram')

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```
